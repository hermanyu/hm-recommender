# 05-1 Preparing Image Data

This notebook is a short bit of code used to copy and resize the image data. The purpose of this resizing is to properly fit the VGG-16 model in the next notebook.

```python
import torch
from torchvision.io import read_image
from torchvision import transforms
from torchvision.utils import save_image

import os
import pandas as pd
import numpy as np
```

```python
articles = pd.read_csv('../data/articles.csv')
```

```python
product_codes = articles['product_code'].drop_duplicates()

product_codes
```

```python
product_codes = list(product_codes)

len(product_codes)
```

```python
base_path = '../images/'

image_directories = os.listdir(base_path)


for folder in image_directories:
    path = base_path+folder
    folder_files = os.listdir(path)
    os.mkdir('../images_unique/'+folder)
    print('folder '+folder+' created')
```

```python
def get_image_path(prod_code):
    prod_string = '0'+str(prod_code)
    path = '../images/'+prod_string[0:3]
    dir_list = os.listdir(path)
    for file_name in dir_list:
        if prod_string in file_name[0:7]:
            return path+'/'+file_name
```

```python
resizer = transforms.Compose([transforms.Resize((240,160))])

for prod in product_codes:
    prod_code = '0'+str(prod)
    folder_path = '../images_unique/'+prod_code[0:3]
    file_path = get_image_path(prod)
    

    try:
        img = read_image(file_path)
    
        if img.shape[1]<img.shape[2]:
            img = torch.transpose(img, 1, 2)
        
        img = resizer(img)
        img = torch.div(img, 255)
        save_image(img, folder_path+'/'+prod_code+'.jpg')
        
    except:
        print("image not found at path: ", prod)
```
